```python
df2 = EWG_Scrape_df.copy()
```

```python
df2
```

```python
df2.dropna(subset=['EWG HEALTH GUIDELINE'], inplace=True)
```

```python
df2.isna().sum()
```

```python
df2.columns
```

```python
df2['Units'] = df2['Utility Measuremnt'].apply(lambda x: x.split()[-1])
df2['Utility Measuremnt'] = df2['Utility Measuremnt'].apply(lambda x: x.split()[0])
df2['EWG HEALTH GUIDELINE'] = df2['EWG HEALTH GUIDELINE'].apply(lambda x: x.split()[0])
```

```python
df2
```

```python
df2.replace(',','', regex=True, inplace=True)
```

```python
df2['Utility Measuremnt'] = pd.to_numeric(df2['Utility Measuremnt'])
df2['EWG HEALTH GUIDELINE'] = pd.to_numeric(df2['EWG HEALTH GUIDELINE'])

```

```python
df2
```

```python
df2['Contaminant_Factor'] = df2['Utility Measuremnt']/df2['EWG HEALTH GUIDELINE']
```

```python
df2
```

```python
utlities_list = list(df2.Utility.unique())
```

```python
new_dataset = []
for utility in utlities_list:
    temp_df = df2[df2.Utility == utility]
    cont_factor_sum = round(temp_df.Contaminant_Factor.sum())
    print(f'The Contaminant factor for {utility} is: {cont_factor_sum}')
    new_dataset_dict = {
    'Utility' : utility,
    'Contaminant_Factor' : cont_factor_sum
    }
    new_dataset.append(new_dataset_dict)
```

```python
new_dataset_df = pd.DataFrame(new_dataset)
```
