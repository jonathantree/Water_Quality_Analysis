```python

for state in states:
    page="https://www.ewg.org/tapwater/state.php?stab=" + str(state)
    print(page)
```

```python
master_df = pd.DataFrame()
for state in states:
    page="https://www.ewg.org/tapwater/state.php?stab=" + str(state)
    url = requests.get(page)
    table = pd.read_html(url.text)
    master_df = master_df.append(table[1],ignore_index=True)
    #input("Press Enter to continue...")
```
