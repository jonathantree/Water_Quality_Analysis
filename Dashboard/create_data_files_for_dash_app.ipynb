```python
# Import Dependencies
import pandas as pd
import numpy as np
import os
import sqlite3
```

### Create one .csv file with census data, contaminant info, and priority

```python
#Files to load
priority_file_to_load = os.path.join("..","Priority_Algo_dev","new_priority_with_fips.csv")
priority_df = pd.read_csv(priority_file_to_load)

# read CSV census plus contaminant data (created in Dash_Practice_jenna.ipynb)
census_cont_df = pd.read_csv("Census_and_cont_outliers_removed.csv")

file_to_read = os.path.join("..","Census_Data_Cleaning","zips_to_counties.csv")
zips_to_counties = pd.read_csv(file_to_read,dtype={"zip": str})

```

```python
priority_df.head()
```

```python
priority_df.columns
```

```python
priority_df.drop(columns=['Unnamed: 0','Geographic_Area_Name', 'County', 'GEOID',
       'Total_Population', 'White', 'Black', 'Native', 'Asian',
       'Pacific_Islander', 'Other', 'Two_or_more_Races', 'Hispanic',
       'Not_Hispanic', 'Not_White', 'pct_White', 'pct_Black', 'pct_Native',
       'pct_Asian', 'pct_Pacific_Islander', 'pct_Other', 'pct_Not_White',
       'pct_Hispanic', 'pct_Not_Hispanic', 'pct_Two_or_more_Races',
       'Simpson_Race_DI', 'Simpson_Ethnic_DI', 'Shannon_Race_DI',
       'Shannon_Ethnic_DI', 'Gini_Index', 'County_FIPS', 'Num_Contaminants',
       'Sum_Population_Served', 'Sum_ContaminantFactor',
       'Min_Contaminant_Factor', 'Max_Contaminant_Factor',
       'Avg_Contaminant_Factor'],inplace=True)
```

```python
priority_df.columns
```

```python
census_cont_df.shape
```

```python
census_cont_priority_df = pd.merge(census_cont_df,priority_df,how='outer',left_on='county_FIPS',right_on='county_FIPS')
```

```python
census_cont_priority_df.head()
```

```python
census_cont_priority_df.count()
```

```python
census_cont_priority_df.sample(30)
```

```python
zips_to_counties.head()
```

```python
df = pd.merge(zips_to_counties,census_cont_priority_df,how='inner',left_on='county_fips',right_on="county_FIPS")
```

```python
df['fips']=df['GEOID'].str[-5:]
```

```python
df.dtypes
```

```python
df.drop(columns=["county_fips","county_FIPS"],inplace=True)
```

```python
df.shape
```

```python
df.columns
```

```python
df[['County','State']] = df['Geographic_Area_Name'].str.split(',',expand=True)
```

```python
df.head()
```

```python
df.columns
```

```python
col = ['zip', 'fips', 'State', 
       'state_id', 'County', 'GEOID','Total_Population', 
       'Simpson_Race_DI', 'Simpson_Ethnic_DI', 'Shannon_Race_DI',
       'Shannon_Ethnic_DI', 'Gini_Index', 'Num_Contaminants',
       'Sum_Population_Served', 'Sum_ContaminantFactor',
       'Avg_Contaminant_Factor', 'Priority']

census_cont_priority_df = df.loc[:,col]
```

```python
census_cont_priority_df.head()
```

```python
census_cont_priority_df.rename(columns = {
        'state_id':'State ID',
        'Total_Population':"County Population",
        'Simpson_Race_DI':'Simpson Race Diversity Index', 
        'Simpson_Ethnic_DI':'Simpson Ethnic Diversity Index', 
        'Shannon_Race_DI':'Shannon Race Diversity Index',
        'Shannon_Ethnic_DI':'Shannon Ethnic Diversity Index', 
        'Gini_Index':"Gini Index", 
        'Num_Contaminants':'Number of Contaminants',
        'Sum_Population_Served':'Population Served', 
        'Sum_ContaminantFactor':'Total Contaminant Factor',
        'Avg_Contaminant_Factor':'Average Contaminant Factor'
    }, inplace=True
)
```

```python
census_cont_priority_df.head()
```

```python
census_cont_priority_df.shape
```

```python
# export to csv. 
census_cont_priority_df.to_csv("census_contaminant_priority_by_zip.csv", index = False)
```
