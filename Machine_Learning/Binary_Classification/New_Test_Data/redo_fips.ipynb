```python
import sqlite3
import pandas as pd
```

```python
db = 'Test_Data_DB.db'
# Connect to SQLite database
conn = sqlite3.connect(db)
  
# Create cursor object
cursor = conn.cursor()
```

```python
sql_query = """SELECT name FROM sqlite_master  
  WHERE type='table';"""
cursor.execute(sql_query)
print(cursor.fetchall())
```

```python
df = pd.read_sql_query("SELECT * FROM Contaminant_Summary",conn)
```
