```python
%matplotlib inline

# Import Dependencies
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python

#Files to load
oregon_data_to_load = os.path.join("..","Resources","Data","Oregon_OR.csv")
hh_data_to_load = os.path.join("..","Resources","Data","asecpub21csv","hhpub21.csv")
```

```python
# Read CSV file into DataFrame
oregon_df = pd.read_csv(oregon_data_to_load,low_memory=False)
oregon_df
```

```python
# Select only the rows corresponding to county data (SUMLEV = 50). There are 36 counties in oregon so we should see 36 rows.
oregon_counties_df = oregon_df.loc[oregon_df["SUMLEV"]==50]
oregon_counties_df
```

```python
#We are interested in the keeping columns relating to Population, Housing Count, 
# "NAME":"County Name"
# GEOID - code with last 5 digits equal to state fips + 0(0) + county fips
# "P0010001":"Total Population"
# "P0010003": "White alone"
# "P0010004": Black alone
# "P0010005": American Indian and Alaska Native Alone
# "P0010006": Asian Alone
# "P0010007": Native Hawaiian and Other Pacific Islander alone
# "P0010008":  Some other race alone
# "P0010009":"Population of 2 or more races"
# "P0020002":"Total Population: Hispanic Or Latino"
# "P0020003":"Total Population: Not Hispanic or Latino"

# See more: https://www2.census.gov/census_2010/01-Redistricting_File--PL_94-171/0FILE_STRUCTURE.pdf

col=["NAME","GEOCODE","P0010001","P0010003","P0010004","P0010005","P0010006",
                "P0010007","P0010008","P0010009","P0020002","P0020003"]

oregon_counties_pop_df = oregon_counties_df.loc[:,col]
oregon_counties_pop_df.head()
```

```python
# Calculate non-white population
oregon_counties_pop_df["Non-White"] = oregon_counties_pop_df[["P0010004","P0010005","P0010006","P0010007","P0010008","P0010009"]].sum(axis=1)
```

```python
oregon_counties_pop_df.drop(columns=["P0010004","P0010005","P0010006","P0010007","P0010008","P0010009"],inplace=True)
oregon_counties_pop_df
```

```python
# Rename the columns
# "P0010001":"Total Population"
# "P0010002" - Population of one race
# "P0010003": "White alone"
# "P0010004": Black alone
# "P0010005": American Indian and Alaska Native Alone
# "P0010006": Asian Alone
# "P0010007": Native Hawaiian and Other Pacific Islander alone
# "P0010008":  Some other race alone
# "P0010009":"Population of 2 or more races"
# "P0020002":"Total Population: Hispanic Or Latino"
# "P0020003":"Total Population: Not Hispanic or Latino"
# "P0020005": Total Population: Not Hispanic or Latino: White alone
# "P0020006":  Total Population: Not Hispanic or Latino: Black or African American alone
# "P0020007":  Total Population: Not Hispanic or Latino: American Indian and Alaska Native alone
# "P0020008":  Total Population: Not Hispanic or Latino: Asian alone
# "P0020009":  Total Population: Not Hispanic or Latino: Native Hawaiian and Other Pacific Islander alone 
# "P0020010": Total Population: Not Hispanic or Latino: Some other race alone
# "P0020011": Total Population: Not Hispanic or Latino: Population of two or more races
oregon_counties_pop_df.rename(columns={"NAME": "County",
                                        "P0010001":"Total Population",
                                        "P0010003":"White",
                                        "P0020002":"Hispanic",
                                        "P0020003":"Not Hispanic",
                                        },inplace=True)
#oregon_counties_pop_df
```

```python
# oregon_counties_pop_df = oregon_counties_pop_df[["County","GEOCODE","Total Population","White","Non-White","Hispanic","Not Hispanic"]]
# oregon_counties_pop_df
```

```python
# Add columns for percentage of each race/ethnicity
oregon_counties_pop_df["% White"] = oregon_counties_pop_df["White"]/oregon_counties_pop_df["Total Population"]
oregon_counties_pop_df["% Non-White"] = oregon_counties_pop_df["Non-White"]/oregon_counties_pop_df["Total Population"]
oregon_counties_pop_df["% Hispanic"] = oregon_counties_pop_df["Hispanic"]/oregon_counties_pop_df["Total Population"]
oregon_counties_pop_df["% Not Hispanic"] = oregon_counties_pop_df["Not Hispanic"]/oregon_counties_pop_df["Total Population"]
```

## Clean Income Data for Oregon

```python
hh_df = pd.read_csv(hh_data_to_load)
```

```python
#pick out the columns we need
# HHINC - househld income category (0-41)
# 1=UNDER $2,500
# 2=$2,500 TO $4,999
# 3=$5,000 TO $7,499
# 4=$7,500 TO $9,999
# 5=$10,000 TO $12,499
# 6=$12,500 TO $14,999
# 7=$15,000 TO $17,499
# 8=$17,500 TO $19,999
# 9=$20,000 TO $22,499
# 10=$22,500 TO $24,999
# 11=$25,000 TO $27,499
# 12=$27,500 TO $29,999
# 13=$30,000 TO $32,499
# 14=$32,500 TO $34,999
# 15=$35,000 TO $37,499
# 16=$37,500 TO $39,999
# 17=$40,000 TO $42,499
# 18=$42,500 TO $44,999
# 19=$45,000 TO $47,499
# 20=$47,500 TO $49,999
# 21=$50,000 TO $52,499
# 22=$52,500 TO $54,999
# 23=$55,000 TO $57,499
# 24=$57,500 TO $59,999
# 25=$60,000 TO $62,499
# 26=$62,500 TO $64,999
# 27=$65,000 TO $67,499
# 28=$67,500 TO $69,999
# 29=$70,000 TO $72,499
# 30=$72,500 TO $74,999
# 31=$75,000 TO $77,499
# 32=$77,500 TO $79,999
# 33=$80,000 TO $82,499
# 34=$82,500 TO $84,999
# 35=$85,000 TO $87,499
# 36=$87,500 TO $89,999
# 37=$90,000 TO $92,499
# 38=$92,500 TO $94,999
# 39=$95,000 TO $97,499
# 40=$97,500 TO $99,999
# 41=$100,000 AND OVER
# GESTFIPS - state FIPS code
# GTCO - FIPS code for county
# HTOTVAL - householdincome VALUE

household_df = hh_df.loc[:,["HHINC","GESTFIPS","GTCO","HTOTVAL"]]
household_df
```

```python
# Rename the columns
household_df.rename(columns={
    "HHINC":"Household Income Category",
    "GESTFIPS":"State FIPS",
    "GTCO":"County Fips",
    "HTOTVAL":"Household Income $"
},inplace=True)
household_df.sample(n=10)
```

```python
# Show Oregon data only
household_income_oregon = household_df.loc[household_df["State FIPS"]==41]
household_income_oregon
```

```python
# Filter on Lane County (the only data for Oregon)
household_income_lane_oregon = household_income_oregon.loc[household_income_oregon["County Fips"]==39]
household_income_lane_oregon
```

```python
hhi_lane = household_income_lane_oregon["Household Income $"].median()
```
