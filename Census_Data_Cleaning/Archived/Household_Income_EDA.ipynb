```python
# Import Dependencies
import pandas as pd
import os
```

```python
county_fips_dictionary={}
with open("State_County_FIPS.txt") as file:
    line_list=[4,5,6]
    for pos,line in enumerate(file):
        # if pos >= 5:
            # fips = list(line.strip().split("        "))
            # print(fips)
        (key,value) = line.strip().split("        ")
        county_fips_dictionary[key] = value

    print("text file to county_fips_dictionary complete")
    
    first_value = list(county_fips_dictionary.values())[0]
    first_key = list(county_fips_dictionary.keys())[0]

    print('First Value: ', first_value)
    print('First Key: ', first_key)
```

```python
#Files to load
data_to_load = os.path.join("..","Resources","Data","asecpub21csv","hhpub21.csv")

hh_df = pd.read_csv(data_to_load)
hh_df
```

```python
#pick out the columns we need
# HHINC - househld income category (0-41)
# 1=UNDER $2,500
# 2=$2,500 TO $4,999
# 3=$5,000 TO $7,499
# 4=$7,500 TO $9,999
# 5=$10,000 TO $12,499
# 6=$12,500 TO $14,999
# 7=$15,000 TO $17,499
# 8=$17,500 TO $19,999
# 9=$20,000 TO $22,499
# 10=$22,500 TO $24,999
# 11=$25,000 TO $27,499
# 12=$27,500 TO $29,999
# 13=$30,000 TO $32,499
# 14=$32,500 TO $34,999
# 15=$35,000 TO $37,499
# 16=$37,500 TO $39,999
# 17=$40,000 TO $42,499
# 18=$42,500 TO $44,999
# 19=$45,000 TO $47,499
# 20=$47,500 TO $49,999
# 21=$50,000 TO $52,499
# 22=$52,500 TO $54,999
# 23=$55,000 TO $57,499
# 24=$57,500 TO $59,999
# 25=$60,000 TO $62,499
# 26=$62,500 TO $64,999
# 27=$65,000 TO $67,499
# 28=$67,500 TO $69,999
# 29=$70,000 TO $72,499
# 30=$72,500 TO $74,999
# 31=$75,000 TO $77,499
# 32=$77,500 TO $79,999
# 33=$80,000 TO $82,499
# 34=$82,500 TO $84,999
# 35=$85,000 TO $87,499
# 36=$87,500 TO $89,999
# 37=$90,000 TO $92,499
# 38=$92,500 TO $94,999
# 39=$95,000 TO $97,499
# 40=$97,500 TO $99,999
# 41=$100,000 AND OVER
# GESTFIPS - state FIPS code
# GTCO - FIPS code for county
# HTOTVAL - householdincome VALUE
# HINTVAL - interest income

household_df = hh_df.loc[:,["HHINC","GESTFIPS","GTCO","HTOTVAL","HINTVAL"]]
household_df
```

```python
household_df["State/County FIPS"] = 1000*household_df["GESTFIPS"] + household_df["GTCO"]
```

```python

household_df.rename(columns={
    "HHINC":"Household Income Category",
    "GESTFIPS":"State FIPS",
    "GTCO":"County Fips",
    "HTOTVAL":"Household Income $",
    "HINTVAL":"Interest Income"
},inplace=True)
household_df.sample(n=10)
```
